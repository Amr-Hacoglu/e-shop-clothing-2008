```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/e-shop-clothing-2008/e-shop clothing 2008 - e-shop clothing 2008.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
# Count plot for main category
plt.figure(figsize=(10, 6))
sns.countplot(x='page 1 (main category)', data=df)
plt.title('Count of Main Categories')
plt.show()
```

```python
# Count plot for colors
plt.figure(figsize=(10, 6))
sns.countplot(x='colour', data=df)
plt.title('Count of Colors')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count plot for model photography
plt.figure(figsize=(10, 6))
sns.countplot(x='model photography', data=df)
plt.title('Count of Model Photography')
plt.show()
```

```python
# Count plot for countries
plt.figure(figsize=(10, 6))
sns.countplot(x='country', data=df)
plt.title('Count of Countries')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution plot for price
plt.figure(figsize=(10, 6))
sns.distplot(df['price'], bins=20)
plt.title('Distribution of Prices')
plt.show()
```
